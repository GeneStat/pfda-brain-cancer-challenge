# Challenge solution 

## 1) import datasets and libraries

```python

#import libraries
import pandas as pd
from sklearn import svm
from sklearn import metrics

#datasets
y= pd.read_csv('data/sc1_Phase1_GE_Outcome.tsv')
x_1 = pd.read_csv('data/sc1_Phase1_GE_Phenotype.tsv')
x_2=pd.read_csv('data/sc1_Phase1_GE_FeatureMatrix.tsv')

```

## 2) Exploring Data

```python

print(y.describe)
print(x_1.describe)
print(x_2.describe)
```

## Create SVM

```python
#Create a svm Classifier
clf = svm.SVC(kernel='linear') # Linear Kernel

#Train the model using the training sets
clf.fit(x_1, y)

#Predict the response for test dataset
y_pred = clf.predict(x_2)


print("Accuracy:",metrics.accuracy_score(y, y_pred))

```
